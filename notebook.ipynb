```python
import puzzle

import pandas as pd
import matplotlib.pyplot as plt

```

```python
pz = puzzle.Puzzle()
df=pd.DataFrame()
```

```python
pz.create_puzzle()
```

```python
tl, tr, bl, br = pz.sort_back_of_puzzle()
print(tl.size)
print(tr.size)
print(bl.size)
print(br.size)
```

```python
test_times = []
solve_iteration = 0
while solve_iteration < 1000:
    pz.solution_time = 0
    test_times.append(pz.random_solve())
    solve_iteration +=1
    if solve_iteration%100==0:
        print(f"solved puzzle {solve_iteration} times")
df['random_grab']=test_times
```

```python
test_times = []
solve_iteration = 0
while solve_iteration < 1000:
    pz.create_puzzle()
    tl, tr, bl, br = pz.sort_back_of_puzzle()
    for area in [tl, tr, bl, br]:
        pz.random_solve(piece_pool=area)
    test_times.append(pz.solution_time)
    solve_iteration +=1
    if solve_iteration%100==0:
        print(f"solved puzzle {solve_iteration} times")
df['area_sort']=test_times
```

```python
fig, ax = plt.subplots()
ax.hist(df['random_grab']/(3600*24), 50,label='random grab')
ax.hist(df['area_sort']/(3600*24), 50,label='area sort')
ax.set_xlabel('solution time [days]')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df['random_grab']/(3600*24), 100,label='random grab')
ax.set_xlabel('solution time [days]')
plt.legend()
plt.show()
```
