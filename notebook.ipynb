```python
test_times = []
solve_iteration = 0
while solve_iteration < 1:
    test_times.append(pz.random_solve()/60)
    solve_iteration +=1
    print(f"solved puzzle {solve_iteration} times")
```

```python
np.savetxt("basic_output", test_times,delimiter=',')
```

```python
import csv

temp = []

with open('basic_output.txt', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        temp.append(float(row[0]))

data = {'random select solution': temp}

df = pd.DataFrame(data=data)
```

```python
fig, ax = plt.subplots()
ax.hist(df['random select solution']/3600, 50)
fig.xlable='solution time [hours]'
plt.show()
```
